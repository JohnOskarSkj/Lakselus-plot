```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.express as px
import dash_table
from dash_table import DataTable, FormatTemplate
import pandas as pd
from dash.dependencies import Input, Output
lakselus = pd.read_csv('lakselus2022-18.csv')
tiltak = pd.read_csv('tiltak2.csv')
logo_link = 'https://upload.wikimedia.org/wikipedia/commons/3/39/Salmo_salar.jpg'
major_categories = list(lakselus['Navn'].unique())
large_tb = lakselus.groupby(['Navn'])['Voksne hunnlus'].agg(['sum', 'count', 'mean', 'median']).reset_index().rename(columns={'count':'Voksne hunnlus', 'sum':'Totalt antall lakselus', 'mean':'Gjennomsnittlig lus', 'median':'Median lakselus'})
ecom_country = lakselus.groupby('Navn')['Voksne hunnlus'].agg('sum').reset_index(name='Voksne hunnlus')
bar_fig_country = px.bar(ecom_country, x='Voksne hunnlus', y='Navn', width=950, height=450, title='(Hold markøren for å filtrere søylediagrammet!)', custom_data=['Navn'], color='Navn', color_discrete_map={'NRS FARMING AS':'lightblue', 'GRIEG SEAFOOD FINNMARK AS':'orange', 'LERØY AURORA AS':'darkblue', 'MIDT-NORSK HAVBRUK AS':'green', 'MOWI ASA':'red', 'SALMAR FARMING AS':'purple'})


app = dash.Dash(__name__)

app.layout = html.Div([
    html.Img(src=logo_link, 
        style={'margin':'30px 0px 0px 0px', 'height':'25%', 'width':'25%' }),
    html.H1('Oversikt over lus'),
    html.Div(
    children=[
    html.Div(
        children=[
        html.H3('Velg oppdrettsfirma'),
        dcc.Dropdown(id='major_cat_dd',
        options=[{'label':category, 'value':category} for category in major_categories],
            value='NRS FARMING AS',
            style={'width':'200px', 'margin':'0 auto'}),
        html.Br(),
        html.H3('Velg lokalitet'),
        dcc.Dropdown(id='minor_cat_dd',
            style={'width':'200px', 'margin':'0 auto'})
        ],
        style={'width':'240px', 'height':'350', 'display':'inline-block', 'vertical-align':'top', 'border':'1px solid black', 'padding':'20px'}),
        html.Div(children=[
                html.H3(id='chosen_major_cat_title'),
                dcc.Graph(id='scatter_compare')
                ],
                 style={'width':'700px', 'height':'300px','display':'inline-block', 'margin-bottom':'2px'}
                 )
        ]), 
        html.Br(),
        html.Br(),
        html.Br(),
   
    html.Div(children=[

        html.Div(dcc.Graph(id='sales_line'), style={'width':'600px', 'height':'350','display':'inline-block'}),
        html.Div(dcc.Graph(id='tiltak_compare'), style={'width':'600px', 'height':'350','display':'inline-block'}),
        dcc.Graph(id='geo_location'),
        html.Div(dcc.Graph(id='major_cat', figure=bar_fig_country), style={'display':'inline-block'}),
        html.Div(dcc.Graph(id='minor_cat'), style={'display':'inline-block'})
            ],
             style={'width':'1000px', 'height':'650px','display':'inline-block'}
             ),
        ],style={'text-align':'center', 'display':'inline-block', 'width':'100%'}
        )

# Create a callback triggered by selecting a column
@app.callback(
    Output('scatter_compare', 'figure'),
    Input('major_cat_dd', 'value'))

def table_country(major_cat):
    major_cat_title = ''
    lakselus_scatter = lakselus.copy()
    
    if major_cat:
        major_cat_title = major_cat
        lakselus_scatter  = lakselus_scatter[lakselus_scatter['Navn'] == major_cat]
    #lakselus_line = lakselus_line.groupby('Uke')['Voksne hunnlus', 'Over lusegrense uke'].agg('sum').reset_index(name='Voksne hunnlus')
    scatter_fig = px.scatter(lakselus_scatter, x='Uke',  y='Voksne hunnlus', color="Over lusegrense uke", opacity=0.7, title=f'Antall voksne hunnlus: {major_cat_title}', height=350)
    return scatter_fig


@app.callback(
   Output('minor_cat_dd', 'options'),
   Output('chosen_major_cat_title', 'children'),
   Input('major_cat_dd', 'value'))

def update_dd(major_cat_dd):
    major_minor = lakselus[['Navn', 'Lokalitetsnavn']].drop_duplicates()
    relevant_minor = major_minor[major_minor['Navn'] == major_cat_dd]['Lokalitetsnavn'].values.tolist()
   
    minor_options = [dict(label=x, value=x) for x in relevant_minor]
    

    if not major_cat_dd:
        major_cat_dd = 'ALL'
    
    major_cat_title = f'Du har valgt oppdrettsfirmaet : {major_cat_dd}'

    return minor_options, major_cat_title

@app.callback(
    Output('tiltak_compare', 'figure'),
    Input('minor_cat_dd', 'value'))

def tiltak_oversikt(major_cat_tiltak):
    tiltak_title = ''
    tiltak_scatter = tiltak.copy()
    
    if major_cat_tiltak:
        tiltak_title = major_cat_tiltak
        tiltak_scatter  = tiltak_scatter[tiltak_scatter['Lokalitetsnavn'] == major_cat_tiltak]
    #lakselus_line = lakselus_line.groupby('Uke')['Voksne hunnlus', 'Over lusegrense uke'].agg('sum').reset_index(name='Voksne hunnlus')
    scatter_tiltak = px.scatter(tiltak_scatter, x='Uke',  y="Tiltak", color="Virkestoff", opacity=0.7, title=f'Lakselusbehandling for: {tiltak_title}', height=350)
    return scatter_tiltak

@app.callback(
    Output('geo_location', 'figure'),
    Input('minor_cat_dd', 'value'))

def location(minor_cat_geo):
    geo_title = 'All'
    lakselus_geo = lakselus.copy()
    
    if minor_cat_geo:
        geo_title = minor_cat_geo
        lakselus_geo = lakselus_geo[lakselus_geo["Lokalitetsnavn"] == minor_cat_geo]
    
    lakselus_geo_point = px.scatter_geo(lakselus_geo, lat="Lat", lon="Lon", scope='europe', title=f'Lakselusbehandling for: {geo_title}', hover_name="Navn")

    return lakselus_geo_point


@app.callback(
    Output('sales_line', 'figure'),
    Input('minor_cat_dd', 'value'))

def update_line(minor_cat):
    minor_cat_title = 'All'
    lakselus_line = lakselus.copy()
    if minor_cat:
        minor_cat_title = minor_cat
        lakselus_line = lakselus_line[lakselus_line['Lokalitetsnavn'] == minor_cat]
    #lakselus_line = lakselus_line.groupby('Uke')['Voksne hunnlus', 'Over lusegrense uke'].agg('sum').reset_index(name='Voksne hunnlus')
    line_graph = px.scatter(lakselus_line, x='Uke',  y='Voksne hunnlus', color="Over lusegrense uke",  title=f'Antall voksne hunnlus: {minor_cat_title}', height=350)
    
    return line_graph

@app.callback(
    Output('minor_cat', 'figure'),
    Input('major_cat', 'hoverData'))

def update_min_cat_hover(hoverData):
    hover_country = 'NRS FARMING AS'
    
    if hoverData:
        hover_country = hoverData['points'][0]['customdata'][0]

    minor_cat_df = lakselus[lakselus['Navn'] == hover_country]
    minor_cat_agg = minor_cat_df.groupby('Lokalitetsnavn')['Voksne hunnlus'].agg('sum').reset_index(name='Voksne hunnlus')
    ecom_bar_minor_cat = px.bar(minor_cat_agg, x='Voksne hunnlus', y='Lokalitetsnavn', orientation='h', height=1000, width=700,title=f'Lakselus ved forksjellige lokaliteter hos: {hover_country}')
    ecom_bar_minor_cat.update_layout({'yaxis':{'dtick':1, 'categoryorder':'total ascending'}, 'title':{'x':0.5}})

    return ecom_bar_minor_cat


 
if __name__ == '__main__':
    app.run_server(debug=False)

```
